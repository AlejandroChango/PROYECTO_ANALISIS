```python
from bson import ObjectId
from pymongo import MongoClient
from datetime import datetime  
import requests



# Conexión a MongoDB
mongo_client = MongoClient('mongodb://localhost:27017/')
mongo_db = mongo_client['DEPORTES']
mongo_collection = mongo_db['DEPORTES']

# Obtener todos los documentos de la colección en MongoDB
documents = list(mongo_collection.find())

# Transformar los documentos a formato JSON (esto es necesario para CouchDB)
json_documents = []
for doc in documents:
    # Convertir ObjectId a cadena antes de agregar el documento
    doc['_id'] = str(doc['_id'])

    # Convertir campos datetime a cadenas en formato ISO 8601
    for key, value in doc.items():
        if isinstance(value, datetime):
            doc[key] = value.isoformat()

    json_documents.append(doc)

# URL de la base de datos CouchDB
couch_url = 'http://admin:admin@127.0.0.1:5984/'
couch_db_name = 'deportes'

# Crear la base de datos en CouchDB si no existe
requests.put(f'{couch_url}/{couch_db_name}')

# Importar los datos a CouchDB
for doc in json_documents:
    response = requests.post(f'{couch_url}/{couch_db_name}/', json=doc)
    if response.status_code == 201:
        print(f'Documento importado con éxito: {doc["_id"]}')
    else:
        print(f'Error al importar el documento: {doc["_id"]}')

print('Migración completada.')
```

```python
import pandas as pd
from pymongo import MongoClient

ruta_archivo_xls = 'productos.xls'


client = MongoClient('mongodb://localhost:27017/')
db = client['Supermercado'] 
coleccion = db['Supermercado']   


hojas_disponibles = pd.ExcelFile(ruta_archivo_xls).sheet_names


for nombre_hoja in hojas_disponibles:
    data_frame = pd.read_excel(ruta_archivo_xls, sheet_name=nombre_hoja)
    datos = data_frame.to_dict(orient='records')
    coleccion.insert_many(datos)

print('Datos insertados en MongoDB.')
```

```python
import pandas as pd
import xml.etree.ElementTree as ET


ruta_archivo_csv = 'productos.csv'

data_frame = pd.read_csv(ruta_archivo_csv)

nombre_raiz = 'datos'

root = ET.Element(nombre_raiz)

columnas = data_frame.columns

for _, fila in data_frame.iterrows():
    elemento_fila = ET.SubElement(root, 'fila')
    for columna in columnas:
        valor = str(fila[columna])
        ET.SubElement(elemento_fila, columna).text = valor

arbol_xml = ET.ElementTree(root)

ruta_archivo_xml = 'productos.xml'

arbol_xml.write(ruta_archivo_xml, encoding='utf-8', xml_declaration=True)

print(f'Archivo CSV convertido a XML: {ruta_archivo_xml}')
```

```python
import sqlite3
import requests
import uuid

ruta_archivo_sqlite = 'covid19.sqlite'

conexion = sqlite3.connect(ruta_archivo_sqlite)
cursor = conexion.cursor()

nombre_tabla = 'DL_case_info'

query = f"SELECT * FROM {nombre_tabla}"
cursor.execute(query)

datos_json = []
for fila in cursor.fetchall():
    fila_dict = dict(zip([column[0] for column in cursor.description], fila))
    datos_json.append(fila_dict)


couch_url = 'http://admin:admin@127.0.0.1:5984/' 
couch_db_name = 'covid19' 

for doc in datos_json:

    doc["_id"] = str(uuid.uuid4())

    response = requests.post(f'{couch_url}/{couch_db_name}/', json=doc)
    if response.status_code == 201:
        print(f'Documento importado con éxito: {doc["_id"]}')
    else:
        print(f'Error al importar el documento: {doc["_id"]}')
```
